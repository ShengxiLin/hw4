# Function to perform bootstrapping
def bootstrap_statistic(sample, n_iterations, stat_func):
    """
    Generate bootstrap samples and calculate the desired statistic for each sample.
    
    Parameters:
    - sample: The original data sample.
    - n_iterations: Number of bootstrap iterations.
    - stat_func: Function to calculate the statistic of interest (e.g., np.mean, np.median).
    
    Returns:
    - List of calculated statistics from each bootstrap sample.
    """
    bootstrap_stats = []
    for _ in range(n_iterations):
        # Generate a bootstrap sample
        bootstrap_sample = np.random.choice(sample, size=n_size, replace=True)
        # Calculate the statistic (mean, median, etc.) of the bootstrap sample
        bootstrap_stat = stat_func(bootstrap_sample)
        bootstrap_stats.append(bootstrap_stat)
    return bootstrap_stats

# To calculate the 95% CI for the population mean:
boot_means = bootstrap_statistic(sample, n_iterations, np.mean)

# Calculate the 95% confidence interval for the mean
lower_bound_mean = np.percentile(boot_means, 2.5)
upper_bound_mean = np.percentile(boot_means, 97.5)

print(f"95% Bootstrap Confidence Interval for the mean: ({lower_bound_mean}, {upper_bound_mean})")

# If you want to calculate the 95% CI for a different statistic, such as the median:

Bootstrapping

# Calculate the difference in health scores (Final - Initial)
df['HealthScoreDiff'] = df['FinalHealthScore'] - df['InitialHealthScore']

# Number of bootstrap samples
n_iterations = 1000

# Function to perform bootstrapping on the differences
def bootstrap_means(data, n_iterations):
    boot_means = []
    for _ in range(n_iterations):
        # Generate a bootstrap sample with replacement
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        # Calculate the mean of the bootstrap sample
        boot_means.append(np.mean(bootstrap_sample))
    return boot_means

# Generate bootstrap means for the health score differences
boot_means = bootstrap_means(df['HealthScoreDiff'], n_iterations)

# Calculate the 95% confidence interval for the mean difference
lower_bound = np.percentile(boot_means, 2.5)
upper_bound = np.percentile(boot_means, 97.5)

print(f"95% Bootstrap Confidence Interval for the mean difference: ({lower_bound}, {upper_bound})")



# Plot the distribution of the bootstrap means
plt.figure(figsize=(8, 5))
plt.hist(boot_means, bins=30, alpha=0.7, color='cyan', edgecolor='black')
plt.axvline(lower_bound, color='red', linestyle='--', label='95% CI Lower Bound')
plt.axvline(upper_bound, color='red', linestyle='--', label='95% CI Upper Bound')
plt.axvline(np.mean(df['HealthScoreDiff']), color='blue', linestyle='-', label='Sample Mean Difference')
